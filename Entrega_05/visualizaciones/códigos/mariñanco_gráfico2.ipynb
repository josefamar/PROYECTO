```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv("/content/gráfica2.csv")
```

```python
df
```

```python
chart = alt.Chart(df).mark_bar().encode(
    x='Total:Q',
    y='Comuna:O',
    color=alt.Color('Cruce:N',
                    scale=alt.Scale(
                        domain=['Auto', 'Personas', "Autos y Personas"],
                        range=['#5A75FA', '#FA945D', "#77C75D"]
                    ),
                    title="Tipos de cruces"),
    tooltip=['Total', "Región", "Índice de Ruralidad por Comuna", "Densidad Poblacional por KM Cuadrado"]
).properties(
    width=300,
    height=400,
    title='Tipos de cruces irregulares'
)
chart.save('gráfica2.html')
```
