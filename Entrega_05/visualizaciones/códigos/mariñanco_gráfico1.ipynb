```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv("/content/Datos_U1 (2).csv")
```

```python
df
```

```python
alt.Chart(df).mark_bar().encode(
    x='Total:Q',
    y='Comuna:O',
    color=alt.Color('Regulación:N',
                    scale=alt.Scale(
                        domain=['Regulado', 'Irregular'],
                        range=['#5A75FA', '#FA945D']
                    ),
                    title="Regulación"),
    tooltip=['Regulación', 'Total', "Región", "Índice de Ruralidad por Comuna", "Densidad Poblacional por KM Cuadrado"]
).properties(
    width=400,
    height=500,
    title='Cantidad de cruces ferroviarios en uso por regulación y comuna'
)
```

```python
chart.save('Datos_U1 (2).html')
```
