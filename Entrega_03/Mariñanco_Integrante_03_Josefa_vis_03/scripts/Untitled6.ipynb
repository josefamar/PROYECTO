```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("/content/Datos_unificados.csv.csv")
```

```python
df
```

```python
def normaliza_regulacion(valor):
    if pd.isna(valor):
        return 'Desconocido'
    v = str(valor).strip().lower()
    if 'irreg' in v:
        return 'Irregular'
    if 'regula' in v:
        return 'Regulado'
    return 'Desconocido'

def normaliza_comuna(valor):
    if pd.isna(valor):
        return 'Desconocido'
    v = str(valor).strip().lower()
    if 'buin' in v:
        return 'Buín'
    return valor.strip()

def normaliza_estado_actual(valor):
    if pd.isna(valor):
        return 'Desconocido'
    v = str(valor).strip().lower()
    if 'en uso' in v:
        return 'En Uso'
    if 'clausurado' in v:
        return 'Clausurado'
    return 'Desconocido'

df['Regulación'] = df['Regulación'].apply(normaliza_regulacion)
df['Comuna'] = df['Comuna'].apply(normaliza_comuna)
df['Estado actual'] = df['Estado actual'].apply(normaliza_estado_actual)
```

```python
conteo = df.groupby(['Comuna', 'Regulación']).size().reset_index(name='Cantidad')
```

```python
chart = alt.Chart(conteo).mark_bar().encode(
    x=alt.X('Regulación:N', title='Tipo de cruce'),
    y=alt.Y('Cantidad:Q', title='Cantidad de cruces'),
    color=alt.Color('Regulación:N', title='Regulación'),
    column=alt.Column('Comuna:N', title='Comuna'),
    tooltip=['Comuna', 'Regulación', 'Cantidad']
).properties(
    width=80,
    height=350,
    title='Cantidad de cruces ferroviarios regulados e irregulares por comuna'
)
display(chart)
```

```python
# Group by Comuna, Regulación, and Estado actual and count the occurrences
conteo_estado = df.groupby(['Comuna', 'Regulación', 'Estado actual']).size().reset_index(name='Cantidad')

# Create a new chart including Estado actual
chart_estado = alt.Chart(conteo_estado).mark_bar().encode(
    x=alt.X('Regulación:N', title='Tipo de cruce'),
    y=alt.Y('Cantidad:Q', title='Cantidad de cruces'),
    color=alt.Color('Estado actual:N', title='Estado Actual'),
    column=alt.Column('Comuna:N', title='Comuna'),
    tooltip=['Comuna', 'Regulación', 'Estado actual', 'Cantidad']
).properties(
    width=80,
    height=350,
    title='Cantidad de cruces ferroviarios por regulación, estado actual y comuna'
)

# Display the new chart
display(chart_estado)
```
